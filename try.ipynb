```python
import json
import pandas as pd

path = r"C:\Users\Mykola\Downloads\senzing.json"

def fix_text(s):
    if s is None:
        return None
    try:
        return s.encode('latin1').decode('utf-8')
    except:
        return s

rows = []

with open(path, 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        obj = json.loads(line)

        # --- name ---
        names = obj.get("NAMES", [])
        name = None
        for n in names:
            if n.get("NAME_TYPE") == "PRIMARY":
                name = fix_text(n.get("NAME_ORG"))
                break

        # --- address ---
        addresses = obj.get("ADDRESSES", [])
        address = fix_text(addresses[0]["ADDR_FULL"]) if addresses else None

        # --- edrpou ---
        identifiers = obj.get("IDENTIFIERS", [])
        edrpou = None
        for ident in identifiers:
            if "NATIONAL_ID_NUMBER" in ident:
                edrpou = ident["NATIONAL_ID_NUMBER"]
                break

        rows.append({
            "id": obj.get("RECORD_ID"),
            "name": name,
            "address": address,
            "edrpou": edrpou
        })

        if i >= 1000:
            break

df = pd.DataFrame(rows)
df.head()
```

```python
import re
import json
import pandas as pd


UA_MOBILE_PREFIXES = {
    "039", "050", "063", "066", "067", "068", "073",
    "091", "092", "093", "094", "095", "096", "097", "098", "099"
}

# Основні валідні коди міст / областей України
UA_LANDLINE_CODES = {
    "0312", "0322", "0332", "0342", "0352", "0362", "0372", "0382",
    "0412", "042", "0432", "044", "045", "0462", "0472", "048", "049",
    "0512", "0522", "0532", "0542", "0552", "056", "057", "061", "062",
    "0642", "0652", "0692"
}


def parse_phones(value):
    """
    Парсить і нормалізує українські номери телефонів.

    Повертає рядок виду:
    +38 (067) 123-45-67; +38 (044) 123-45-67

    Некоректні або підозрілі номери ігнорує.
    """

    if pd.isna(value) or str(value).strip() == "":
        return ""

    raw_value = str(value).strip()
    raw_phones = []

    # Якщо всередині JSON
    try:
        data = json.loads(raw_value)

        if isinstance(data, dict):
            for key in ["phone", "phones", "contacts", "contact_phones"]:
                if key in data:
                    if isinstance(data[key], list):
                        raw_phones.extend(map(str, data[key]))
                    else:
                        raw_phones.append(str(data[key]))
        elif isinstance(data, list):
            raw_phones.extend(map(str, data))
        else:
            raw_phones.append(raw_value)

    except Exception:
        raw_phones = re.split(r"[;,|/\n]+", raw_value)

    result = []
    seen = set()

    for raw_phone in raw_phones:
        digits = re.sub(r"\D+", "", str(raw_phone))

        if not digits:
            continue

        # Прибираємо явне сміття
        if len(set(digits)) <= 2:
            continue

        if digits.startswith("000") or digits.startswith("001"):
            continue

        # +380XXXXXXXXX або 380XXXXXXXXX
        if digits.startswith("380") and len(digits) == 12:
            national = "0" + digits[3:]

        # 80XXXXXXXXX
        elif digits.startswith("80") and len(digits) == 11:
            national = "0" + digits[2:]

        # 0XXXXXXXXX
        elif digits.startswith("0") and len(digits) == 10:
            national = digits

        else:
            continue

        formatted = None

        # Мобільний номер
        if national[:3] in UA_MOBILE_PREFIXES:
            formatted = (
                f"+38 ({national[:3]}) "
                f"{national[3:6]}-{national[6:8]}-{national[8:10]}"
            )

        # Міський номер з 3-значним кодом, наприклад 044, 048, 056
        elif national[:3] in UA_LANDLINE_CODES:
            formatted = (
                f"+38 ({national[:3]}) "
                f"{national[3:6]}-{national[6:8]}-{national[8:10]}"
            )

        # Міський номер з 4-значним кодом, наприклад 0342, 0322
        elif national[:4] in UA_LANDLINE_CODES:
            formatted = (
                f"+38 ({national[:4]}) "
                f"{national[4:6]}-{national[6:8]}-{national[8:10]}"
            )

        # Все інше вважаємо підозрілим і не додаємо
        else:
            continue

        if formatted and formatted not in seen:
            seen.add(formatted)
            result.append(formatted)

    return "; ".join(result)

df=pd.read_csv(r"C:\OTP\company_results.csv")


df["phones_clean"] = (
    df["phone"].fillna("").astype(str) + "; " +
    df["contacts_json"].fillna("").astype(str)
).apply(parse_phones)
```

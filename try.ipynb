```python
import json
import pandas as pd

path = r"C:\Users\Mykola\Downloads\senzing.json"

def fix_text(s):
    if s is None:
        return None
    try:
        return s.encode('latin1').decode('utf-8')
    except:
        return s

rows = []

with open(path, 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        obj = json.loads(line)

        # --- name ---
        names = obj.get("NAMES", [])
        name = None
        for n in names:
            if n.get("NAME_TYPE") == "PRIMARY":
                name = fix_text(n.get("NAME_ORG"))
                break

        # --- address ---
        addresses = obj.get("ADDRESSES", [])
        address = fix_text(addresses[0]["ADDR_FULL"]) if addresses else None

        # --- edrpou ---
        identifiers = obj.get("IDENTIFIERS", [])
        edrpou = None
        for ident in identifiers:
            if "NATIONAL_ID_NUMBER" in ident:
                edrpou = ident["NATIONAL_ID_NUMBER"]
                break

        rows.append({
            "id": obj.get("RECORD_ID"),
            "name": name,
            "address": address,
            "edrpou": edrpou
        })

        if i >= 1000:
            break

df = pd.DataFrame(rows)
df.head()
```

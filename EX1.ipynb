```python
import numpy as np
```

```python
BOARD_SIZE = 3
PLAYER_X = 1
PLAYER_0 = -1
EMPTY = 0
```

```python
board = np.zeros((BOARD_SIZE, BOARD_SIZE))
board
```

```python
np.argwhere(board==EMPTY)
```

```python
def random_moves(board, n_moves):
    player = PLAYER_X

    for _ in range(n_moves):
        empty_cells = np.argwhere(board == EMPTY)
        if empty_cells.size == 0:
            break

        id = np.random.choice(len(empty_cells))
        i, j = empty_cells[id]

        board[i, j] = player
        player = -player

    return board
```

```python
def check_winner(board):
    lines = np.concatenate([
        board.sum(axis=0),
        board.sum(axis=1),
        [np.trace(board)],
        [np.trace(np.fliplr(board))]
    ])

    if 3 in lines:
        return PLAYER_X
    if -3 in lines:
        return PLAYER_0
    return 0
```

```python
board = random_moves(board, n_moves=4)
result = check_winner(board)

board     
result     
```

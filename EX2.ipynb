```python
import numpy as np
import copy
```

```python

a = np.arange(16).reshape(4, 4)

print("a:")
print(a, end="\n\n")
```

```python
b = a
print("b = a:")
print(b, end="\n\n")

b[0, 0] = 14111

print("a:")
print(a)
print("b:")
print(b, end="\n\n")
```

```python
a = np.arange(16).reshape(4, 4)
b = a.view()

print("b = a.view():")
print(b, end="\n\n")

b[0, 1] = 1231

print("a:")
print(a)
print("b:")
print(b, end="\n\n")
```

```python
a = np.arange(16).reshape(4, 4)
b = a.copy()

print("b = a.copy():")
print(b, end="\n\n")

b[0, 2] = 33333

print("a:")
print(a)
print("b:")
print(b, end="\n\n")
```

```python
a = np.array([
    [[1], [2]],
    [[3], [4]]
], dtype=object)

b = a.copy()
c = copy.deepcopy(a)

print(a, end="\n\n")

b[0, 0].append(999)

print("a:")
print(a)
print("b:")
print(b)
print("c (deepcopy):")
print(c)
```
